```python
import numpy as np
from sklearn.datasets import *
from sklearn.utils import shuffle
# Generate isotropic Gaussian blobs for clustering.
X, y = make_blobs(centers=17, n_samples=1000, cluster_std=1.4, n_features=100)
# X, y = make_classification(n_classes=27, n_clusters_per_class=1, n_samples=1000, n_features=20, n_informative=18)
X, y = shuffle(X, y, random_state=0)
np.savetxt('points.txt', X, delimiter='\t')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 4, 4
plt.scatter(X[:, 0], X[:, 1], c=y, s=75)
plt.xlabel("first feature")
plt.ylabel("second feature")
plt.show()
```

```python
X = np.loadtxt('points.txt', delimiter='\t')
```

```python
from sklearn.cluster import *
kmeans = KMeans(n_clusters=7, random_state=100, verbose=False)
labels = kmeans.fit_predict(X)
print(kmeans.inertia_)
print(len(kmeans.cluster_centers_))
print(len(labels))
```

```python
dbs = DBSCAN(eps = 40, min_samples=10)
labels = dbs.fit_predict(X)
print(labels)
print(dbs.labels_)
print(np.max(labels))
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=17)

X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(12,10))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.title('t-SNE projection')
plt.show()
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
X_reduced = pca.fit_transform(X)

print('Projecting %d-dimensional data to 2D' % X.shape[1])

plt.figure(figsize=(12,10))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.title('PCA projection')
plt.show()
```
